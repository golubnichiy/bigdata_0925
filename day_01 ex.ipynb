Создайте новый столбец 'income_category', который будет содержать значения 'low', 'medium' и 'high' в зависимости от уровня дохода ('income').

```python
# Вставьте код
def categorize_income(income):
    if income <= 30:
        return 'low'
    elif income <= 60:
        return 'medium'
    else:
        return 'high'

gapminder_data['income_category'] = gapminder_data['income'].apply(categorize_income)

gapminder_data
```

```python
bins = [0, 50, 75, 100] # диапазоны: 0–50, 51–75, 76–100  
labels = ['Low', 'Average', 'High'] # столбцы  
my_data = gapminder_data['income_category'] = pd.cut(gapminder_data['income'], bins = bins, labels = labels, include_lowest = True)
print (my_data)
```

```python
my_data = gapminder_data['income_category'] = pd.qcut(gapminder_data['income'], labels = labels ,q=3)
print (my_data)
```
