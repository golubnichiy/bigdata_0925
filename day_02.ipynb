```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('HWS.csv')
```

```python
print(data.head())
print(data.shape)
```

### Hist

```python
data['Height'].hist()
plt.show()
```

```python
data['Height'].hist(bins=20)
plt.show()
```

### Bar plot

```python
data_gr_height = data.groupby('Index')['Height'].mean()
```

```python
data_gr_height
```

```python
data_gr_height.plot(kind='bar', title='Срений рост по BMI', rot=0)
plt.show()
```

```python
aq = pd.read_csv('aq.csv')
print(aq)
```

```python
aq.plot(x='Date', y='NO2', rot=30)
plt.show()
```

### Диаграмма рассеяния

```python
data.plot(x='Height', y='Weight', kind='scatter')
plt.show()
```

### Добавление уровней

```python
data[data['Gender'] == 'Male']['Height'].hist(bins=20, alpha=0.6)
data[data['Gender'] == 'Female']['Height'].hist(bins=20, alpha=0.6)
plt.legend(['Male', 'Female'])
plt.show()
```
